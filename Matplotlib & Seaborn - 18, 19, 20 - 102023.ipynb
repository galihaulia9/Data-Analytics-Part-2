### Line Chart

- Using Matplotlib

```python
yield_apples = [0.89, 0.91, 0.92, 0.93, 0.94, 0.95] #ton per hectare
```

```python
plt.plot(yield_apples)
```

```python
years = [2017, 2018, 2019, 2020, 2021, 2022]
yield_apples = [0.89, 0.91, 0.92, 0.93, 0.94, 0.95]
```

```python
plt.plot(years, yield_apples) #plit.plot(x,y)
```

```python
plt.plot(years, yield_apples)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
```

```python
apples = [0.89, 0.91, 0.93, 0.97, 0.986, 0.943]
tangerines = [0.79, 0.61, 0.93, 0.97, 0.962, 0.931]
```

```python
plt.plot(years, yield_apples)
plt.plot(years, tangerines)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
```

```python
plt.plot(years, apples)
plt.plot(years, tangerines)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Tangerines'])
```

```python
plt.plot(years, apples, marker = 'o')
plt.plot(years, tangerines, marker = 'x')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title("Crop Yields in Kanto")
plt.legend(['Apples', 'Tangerines'])
```

```python
plt.figure(figsize=(12,6))

plt.plot(years, tangerines, marker = 'o')
plt.title("Yield of Tangerines (tons per hectare)")
```

```python
import numpy as np

plt.hist(iris_df.sepal_width, bins=np.arange(2, 5, 0.25));
```

```python
# bins of unequal sizes
plt.hist(iris_df.sepal_width, bins=[1,3,4,4.5])
```

```python
setosa_df = iris_df[iris_df.species == 'setosa']
versicolor_df = iris_df[iris_df.species == 'versicolor']
virginica_df = iris_df[iris_df.species == 'virginica']
```

```python
plt.hist(setosa_df.sepal_width, alpha = 0.4)
plt.hist(versicolor_df.sepal_width, alpha = 0.4)
```
