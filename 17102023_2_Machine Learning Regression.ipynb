### DATA SALES TV

This dataset is used to analyze the relationship between advertising budget on TV (TV) and sales (Sales). Simple linear regression analysis can be used to understand the extent to which TV advertising impacts the sales of the advertised product or service. 

First Import data using the required libraries

**Import Libraries**

```python
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

Call and read data from files tvmarketing.csv

```python
# Membaca data
df = pd.read_csv("tvmarketing.csv")
```

Display the first 5 rows

```python
df.head()
```

 Display statistical information 

```python
df.describe()
```

**Checking the info of the data type in the file**

The result of this check is that all columns are clean and there are no blanks (notnull)

```python
df.info()
```

**Separating independent (X) and dependent (y) variables**

```python
X = df[['TV']]
y = df['Sales']
```

**Divide the data into training data and testing data**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=42)
```

**Inisialisasi model regresi linear**

```python
lr = LinearRegression()
```

**Train the model**

```python
lr.fit(X_train, y_train)
```

**Predict and print model evaluation results**

```python
#  Melakukan prediksi
y_pred = lr.predict(X_test)
```

```python
# Mencetak hasil evaluasi model
mse = mean_squared_error(y_test, y_pred)
r_squared = r2_score(y_test, y_pred)
```

```python
print('Mean Squared Error:', mse)
print('R-squared:', r_squared)
```

**Result**

The MSE has a value of about 9.57. The smaller the MSE value, the better the linear regression model performs, as this indicates that the model predictions are closer to the actual values.

The R-squared has a value of about 0.64. This means about 64% of the variability in the data can be explained by the model. The higher the R-squared value, the better the linear regression model performs in explaining the variation in the data.

In a nutshell, these results indicate that the linear regression model is good at predicting the dependent variable (Sales) based on the independent variable (TV).

**Creating a plot Actual vs Predicted**

```python
c = [i for i in range(len(y_test))]
plt.plot(c, y_test, color="blue", linewidth=2, linestyle="-", label="Actual")
plt.plot(c, y_pred, color="red", linewidth=2, linestyle="-", label="Predicted")
plt.title('Actual and Predicted', fontsize=20)
plt.xlabel('Index', fontsize=18)
plt.ylabel('Sales', fontsize=16)
```

The result of plotting actual vs predicted.

The red line is predicted while the blue line is actual. The result of the above query produces a plot where the blue line and the red line are close or almost aligned, meaning that the model accurately predicts the value of the value in 'Sales'. So the conclusion is that the similarity between the model prediction results and the actual data is successful because the location of the red line (predicted) is close or almost aligned with the blue line (actual).

**Creating a Plot Error Terms**

```python
# Indeks berdasarkan panjang y_test (sama dengan panjang asli)
c = [i for i in range(len(y_test))]

# Menghitung error terms (selisih antara y_test dan y_pred)
error_terms = y_test - y_pred

# Membuat plot untuk error terms
plt.plot(c, error_terms, color="blue", linewidth=2, linestyle="-")
plt.title('Error Terms', fontsize=20)
plt.xlabel('Index', fontsize=18)
plt.ylabel('y_test - y_pred', fontsize=16)
plt.show()
```

Blue Line : Shows the difference between the actual value and the predicted value for each data point in the test data. If the line is above zero, it means the model overpredicted, while if it is below zero, it means the model underpredicted. The results indicate that the model has varying errors for each data point in the test data.
