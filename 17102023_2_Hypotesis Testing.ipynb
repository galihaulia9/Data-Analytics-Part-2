```python
import pandas as pd
from scipy import stats
from statsmodels.stats import weightstats as stests
```

```python
df = pd.read_csv('coffee.csv')
```

This is a collection of coffee prices from monthly calculations. This dataset includes variables such as opening, closing, highest, lowest, total revenue and currency value.

```python
df.head()
```

```python
df[['Open', 'Close']].describe()
df.head(15)
```

```python
df.shape
```

- H0 : means difference between two sample is 0
- H1 : means difference between two sample is not 0

```python
ttest,pval = stats.ttest_rel(df['Open'],df['Close'])
print(pval)
```

```python
if pval < 0.15 :  # error tolerance limit of 15%
    print("Reject null hypothesis")
else:
    print("Accept null hypothesis")
```

Conclusion : The test results above that H0 is rejected mean that there is a difference in the two samples or data.
